```python
q = "energy"
jurisdiction ="ny"
page = 1
apikey = "855b666e-b40f-4674-9da0-2cd8eb9c3688"
base_api_link = "https://v3.openstates.org"
```

```python
url = f"{base_api_link}/bills?jurisdiction={jurisdiction}&sort=updated_desc&q={q}&page={page}&per_page=20&apikey={apikey}"
```

```python
page
```

```python
initial_request = requests.get(url)
if initial_request.status_code == 200:
    # get json
    json_response = initial_request.json()
    # get number of pages 
    num_pages = json_response["pagination"]["max_page"]
    # current page 
    current_page = json_response["pagination"]["page"]
    # save the json response from the 1st page 
    
    file_name = f'{jurisdiction}_{q}_{page}.json'
    file_path = os.path.join("Dataset", "openstates_bills", jurisdiction, file_name)
    
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(json_response, f, ensure_ascii=False, indent=4)
        
```

```python
page += 1

while page <= num_pages:
    
    # print which page 
    print(page, num_pages)
    
    updated_url = f"{base_api_link}/bills?jurisdiction={jurisdiction}&sort=updated_desc&q={q}&page={page}&per_page=20&apikey={apikey}"
    print(url)
    
    # make next request
    next_api_request = requests.get(updated_url)
    
    # get json
    json_response = next_api_request.json()
    # filename
    file_name = f'{jurisdiction}_{q}_{page}.json'
    file_path = os.path.join("Dataset", "openstates_bills", jurisdiction, file_name)

    # save json
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(json_response, f, ensure_ascii=False, indent=4)
        
    # increment page 
    page += 1
    
    # Sleep for 5 seconds
    time.sleep(20) 
            
```
